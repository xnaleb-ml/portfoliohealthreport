# Example notebook running the extraction and summarization pipeline

```python
import os
from pathlib import Path

from pipeline import ExtractionPipeline

# Set up environment variables and paths
BASE_DIR = Path.cwd()
# A CLAUDE_API_KEY is required to run the pipeline, set it as an environment variable
CLAUDE_API_KEY = ""

os.environ["ANTHROPIC_API_KEY"] = CLAUDE_API_KEY
os.environ["LLM_MODEL"] = "claude-sonnet-4-6"
os.environ["DEFAULT_LLM_TEMPERATURE"] = "0.0"
os.environ["MAX_TOKENS_TO_GENERATE"] = "10000"
```

# Configuration keys for models and file paths, pipeline settings

```python
# These keys should match the entries in the config.yaml file
extractor_model_config_key = "portfolio_health_reviewer"
summarizer_model_config_key = "portfolio_executive_summarizer"
# Path for the colleagues file containing names and email addresses
colleagues_filepath = BASE_DIR / "Colleagues.txt"
# Path for the input email files unzipped
input_folder = BASE_DIR / "data"
# Path for the output files
output_folder = BASE_DIR / "outputfolder2"
# Path for the anonymized email files
anonymized_folder = output_folder / "anonymized"
# Set maximum concurrency for parallel execution and whether to run in parallel
max_concurrency = 5
run_parallel = True
```

# Initializing pipeline

```python
pipeline = ExtractionPipeline(
    extractor_model_config_key=extractor_model_config_key,
    summarizer_model_config_key=summarizer_model_config_key,
    colleagues_filepath=colleagues_filepath,
    input_folder=input_folder,
    output_folder=output_folder,
    anonymized_folder=anonymized_folder,
    max_concurrency=max_concurrency,
)
```

# How to run pipeline parellel
- 1. The output folder is created
- 2. The emails are anonymized and are saved to the anonimyzed folder within the output folder
- 3. Executive summaries(natural language) are generated per extracted project name, saved within the output folder(should be in a separate folder) 

```python
await pipeline.run(parallel=run_parallel)
```

# How to run pipeline sequentially

```python
await pipeline.run(parallel=False)
```
